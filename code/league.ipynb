```python
import pandas as pd
```

```python
df_origin = pd.read_csv('../data/E0.csv')

df_origin.head(10)
```

```python
print(list(df_origin.columns.tolist()))
```

```python
keepcols = ['Div', 'Date', 'Time', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'HTR', 'Referee', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR']

df = df_origin[keepcols]

df.head()
```

```python
def homepoints(FTR):
    if FTR == 'H':
        return 3
    if FTR == 'D':
        return 1
    if FTR == 'A':
        return 0
    return 0

def awaypoints(FTR):
    if FTR == 'A':
        return 3
    if FTR == 'D':
        return 1
    if FTR == 'H':
        return 0
    return 0


df2 = df.copy()
df2.loc[:, 'HomePoints'] = df.loc[:, 'FTR'].apply(homepoints)
df2.loc[:, ('AwayPoints')] = df.loc[:, ('FTR')].apply(awaypoints)
df2.loc[:, 'Score'] = df['FTHG'].astype(str) + '-' + df['FTAG'].astype(str)

df2.head()
```

```python
df2['Date'] = pd.to_datetime(df2['Date']).dt.date
print(df2['Date'].dtype)
df2.head(10)
```

```python
# df2.pivot_table(index=['HomeTeam'], columns='AwayTeam', values='Score', fill_value='-', aggfunc='first')
```

```python
def homeleague(x):
        results = {
            'HomePoints': x['HomePoints'].sum(),
            'HomeWins': x[x['FTR']== 'H']['HomeTeam'].count(),
            'HomeDraws': x[x['FTR']== 'D']['HomeTeam'].count(),
            'HomeLoss': x[x['FTR']== 'A']['HomeTeam'].count(),
            'HomeGoalsFor': x['FTHG'].sum(),
            'HomeGoalsAgainst': x['FTAG'].sum(),
            'HomeGoalsDiff': x['FTHG'].sum() - x['FTAG'].sum()
        }
        return pd.Series(results)

def awayleague(x):
    results = {
        'AwayPoints': x['AwayPoints'].sum(),
        'AwayWins': x[x['FTR']== 'A']['AwayTeam'].count(),
        'AwayDraws': x[x['FTR']== 'D']['AwayTeam'].count(),
        'AwayLoss': x[x['FTR']== 'H']['AwayTeam'].count(),
        'AwayGoalsFor': x['FTAG'].sum(),
        'AwayGoalsAgainst': x['FTHG'].sum(),
        'AwayGoalsDiff': x['FTAG'].sum() - x['FTHG'].sum()
    }
    return pd.Series(results)

df_homeleague = df2.groupby('HomeTeam').apply(homeleague)
df_awayleague = df2.groupby('AwayTeam').apply(awayleague)
df_league_temp = pd.merge(df_homeleague, df_awayleague, left_index=True, right_index=True)

df_league_temp.head(20)
```

```python


def league(x):
    results = {
        'Points': x['HomePoints'].sum() + x['AwayPoints'].sum(),
        'Wins': x['HomeWins'].sum() + x['AwayWins'].sum(),
        'Draws':  x['HomeDraws'].sum() + x['AwayDraws'].sum(),
        'Loss': x['HomeLoss'].sum() + x['AwayLoss'].sum(),
        'GoalsFor': x['HomeGoalsFor'].sum() + x['AwayGoalsFor'].sum(),
        'GoalsAgainst': x['HomeGoalsAgainst'].sum() + x['AwayGoalsAgainst'].sum(),
        'GoalDifference': x['HomeGoalsDiff'].sum() + x['AwayGoalsDiff'].sum()
    }
    return pd.Series(results)

df_league = df_league_temp.groupby('HomeTeam').apply(league)

df_league.sort_values(by='Points', ascending=False, inplace=True)

df_league.head(20)
```

```python
# Best away results

df_away_results = df_league_temp.loc[:, ('AwayPoints', 'AwayGoalsDiff')]

awaypoints = df_away_results.loc[:, ('AwayPoints')].mean()
#  average away points
print('Average away points: ', awaypoints)

# Average away goal difference
awaygoaldiff = df_away_results.loc[:, ('AwayGoalsDiff')].mean()
print('Average away goal difference: ', awaygoaldiff)

# filter away teams with more than 2 points
df_away_results = df_away_results[(df_away_results['AwayPoints'] > awaypoints) & (df_away_results['AwayGoalsDiff'] > awaygoaldiff)]

df_away_results.sort_values(by='AwayPoints', ascending=False, inplace=True)
df_away_results.head(10)
```

```python
# get last 5 results for home teams
team = 'Newcastle'

df_newcastle = df2.query("(HomeTeam == @team) or (AwayTeam == @team)")

df_newcastle.head(38)
```

```python

def form(x):
    if x['HomeTeam'] == 'Newcastle' and x['FTR'] == 'H':
        return 'W'
    if x['AwayTeam'] == 'Newcastle' and x['FTR'] == 'A':
        return 'W'
    if x['HomeTeam'] == 'Newcastle' and x['FTR'] == 'A':
        return 'L'
    if x['AwayTeam'] == 'Newcastle' and x['FTR'] == 'H':
        return 'L'
    if x['FTR'] == 'D':
        return 'D'
    return 'U'

df_newcastle_form = df_newcastle.loc[:, ('HomeTeam', 'AwayTeam', 'FTR')]
df_newcastle_form.loc[:, 'Form'] = df_newcastle_form.apply(form, axis=1)
df_newcastle_form.reset_index(inplace=True)

df_newcastle_form = df_newcastle_form['Form']
df_newcastle_form.tail(5)
```
