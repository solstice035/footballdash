```python
import os
from dotenv import load_dotenv
import pandas as pd
import numpy as np
import requests
import json

pd.set_option("display.max_columns", 500)

# API key
load_dotenv('../../.env')
api_key = os.environ['APIFOOTBALL_AUTH_KEY']

print(api_key)
```

```python
# get data from api: https://v3.football.api-sports.io/
# premier league (id is 39) data from current season (2022) using GET request
url = 'https://v3.football.api-sports.io/fixtures?league=39&season=2010'

payload = {}
headers = {'x-rapidapi-key': api_key, 'x-rapidapi-host': 'v3.football.api-sports.io'}
response = requests.request("GET", url, headers=headers, data=payload)

# convert json to dataframe
data = json.loads(response.text)
df = pd.json_normalize(data, record_path=['response'], meta=['results'])

print(df.shape)
df.head()
```

```python
# Create a function to get data from api
# https://v3.football.api-sports.io/
#  using GET request
def get_data(league_id, season):
    url = f'https://v3.football.api-sports.io/fixtures?league={league_id}&season={season}'

    payload = {}
    headers = {'x-rapidapi-key': api_key, 'x-rapidapi-host': 'v3.football.api-sports.io'}
    response = requests.request("GET", url, headers=headers, data=payload)

    # convert json to dataframe
    data = json.loads(response.text)
    df = pd.json_normalize(data, record_path=['response'], meta=['results'])
    
    return df

# loop to get data from api for premier league from 2010 to 2022 using concat
df = pd.DataFrame()
for season in range(2010, 2023):
    df = pd.concat([df, get_data(39, season)], axis=0)
    

print(df.shape)
df.head()

```

```python
# export data to csv and pickle and excel to new sub folder "sports_api"
df.to_csv('../../data/sports_api/epl_2010_2022.csv', index=False)
df.to_pickle('../../data/sports_api/epl_2010_2022.pkl')
df.to_excel('../../data/sports_api/epl_2010_2022.xlsx', index=False)
```
