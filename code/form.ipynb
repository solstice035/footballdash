```python
import pandas as pd
```

```python
df = pd.read_pickle('../data/pickle/matches.pkl')
```

```python
# get last 5 results for home teams
team = 'Newcastle'

df_newcastle = df.query("(HomeTeam == @team) or (AwayTeam == @team)")

df_newcastle.head(38)
```

```python
def form(x):
    if x['HomeTeam'] == 'Newcastle' and x['FTR'] == 'H':
        return 'W'
    if x['AwayTeam'] == 'Newcastle' and x['FTR'] == 'A':
        return 'W'
    if x['HomeTeam'] == 'Newcastle' and x['FTR'] == 'A':
        return 'L'
    if x['AwayTeam'] == 'Newcastle' and x['FTR'] == 'H':
        return 'L'
    if x['FTR'] == 'D':
        return 'D'
    return 'U'

df_newcastle_form = df_newcastle.loc[:, ('HomeTeam', 'AwayTeam', 'FTR')]
df_newcastle_form.loc[:, 'Form'] = df_newcastle_form.apply(form, axis=1)
df_newcastle_form.reset_index(inplace=True)

df_newcastle_form = df_newcastle_form['Form']
df_newcastle_form.tail(5)
```
